```python
def max_of_three(a, b, c):
    """Find the maximum of three numbers."""
    return max(a, b, c)


def sum_of_list(numbers):
    """Sum all the numbers in a list."""
    return sum(numbers)


# Test for sum_of_list
sample_list1 = [8, 2, 3, 0, 7]
print("Sum of list:", sum_of_list(sample_list1))


def multiply_list(numbers):
    """Multiply all the numbers in a list."""
    result = 1
    for num in numbers:
        result *= num
    return result


# Test for multiply_list
sample_list2 = [8, 2, 3, -1, 7]
print("Product of list:", multiply_list(sample_list2))


def reverse_string(string):
    """Reverse a given string."""
    return string[::-1]


# Test for reverse_string
sample_string = "1234abcd"
print("Reversed string:", reverse_string(sample_string))


def factorial(n):
    """Calculate the factorial of a non-negative integer."""
    if n == 0:
        return 1
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result


# Test for factorial
print("Factorial of 5:", factorial(5))


def is_in_range(number, start, end):
    """Check if a number falls within a given range."""
    return start <= number <= end


# Test for is_in_range
print("Is 5 in range 1-10:", is_in_range(5, 1, 10))


def count_case(string):
    """Count uppercase and lowercase letters in a string."""
    upper_case = sum(1 for char in string if char.isupper())
    lower_case = sum(1 for char in string if char.islower())
    return upper_case, lower_case


# Test for count_case
sample_string2 = "The quick Brow Fox"
upper, lower = count_case(sample_string2)
print("No. of Upper case characters:", upper)
print("No. of Lower case characters:", lower)


def unique_elements(lst):
    """Return a list with distinct elements."""
    return list(set(lst))


# Test for unique_elements
sample_list3 = [1, 2, 3, 3, 3, 3, 4, 5]
print("Unique List:", unique_elements(sample_list3))


def is_prime(num):
    """Check if a number is prime."""
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True


# Test for is_prime
print("Is 7 prime:", is_prime(7))


def even_numbers(lst):
    """Return even numbers from a given list."""
    return [num for num in lst if num % 2 == 0]


# Test for even_numbers
sample_list4 = [1, 2, 3, 4, 5, 6, 7, 8, 9]
print("Even numbers:", even_numbers(sample_list4))


def is_perfect_number(num):
    """Check if a number is perfect."""
    divisors = [i for i in range(1, num) if num % i == 0]
    return sum(divisors) == num


# Test for is_perfect_number
print("Is 6 a perfect number:", is_perfect_number(6))


def is_palindrome(string):
    """Check if a string is a palindrome."""
    string = string.replace(" ", "").lower()
    return string == string[::-1]


# Test for is_palindrome
sample_string3 = "nurses run"
print("Is palindrome:", is_palindrome(sample_string3))
```
